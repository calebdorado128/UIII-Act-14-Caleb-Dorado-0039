# Problema 15: Una persona adquirió un producto para pagar en 20 meses. El primer mes pagó 10 €, el segundo 20 €, el tercero 40 € y así sucesivamente. Realizar un algoritmo para determinar cuánto debe pagar mensualmente y el total de lo que pagó después de los 20 meses.

```python
# Entrada interactiva de datos [3, 4]
pago_inicial = float(input("Ingrese el pago del primer mes: "))
total_meses = int(input("Ingrese el número total de meses a pagar: "))

# Inicialización de variables de control [2, 6, 8]
pago_mensual = pago_inicial
total_pagado = 0

# Estructura repetitiva para calcular mes a mes [2, 6]
# range(1, total_meses + 1) genera la secuencia de meses solicitada
for mes in range(1, total_meses + 1):
    # Mostrar el pago del mes actual [7, 9]
    print(f"Mes {mes}: {pago_mensual} €")

    # Acumular el pago en el total [6, 8]
    total_pagado = total_pagado + pago_mensual

    # Calcular el pago para el mes siguiente (se duplica) [5]
    pago_mensual = pago_mensual * 2

# Mostrar el resultado final acumulado [7, 10]
print(f"\nEl total pagado tras los {total_meses} meses es: {total_pagado} €")
```
